```python
# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
import os
from dotenv import load_dotenv

load_dotenv()

CONSUMER_KEY = os.getenv('CONSUMER_KEY')
CONSUMER_SECRET = os.getenv('CONSUMER_SECRET')
OAUTH_TOKEN = os.getenv('OAUTH_TOKEN')
OAUTH_TOKEN_SECRET = os.getenv('OAUTH_TOKEN_SECRET')
```

Query Twitter's API for JSON data for each tweet ID in the Twitter archive.

```python
from utils import timer
import json

fails_dict = {}


@timer
def save_tweets_to_json():
    """Save each tweet's returned JSON as a new line in a .txt file"""
    count = 0
    with open('data/tweet_json.txt', 'w') as outfile:
        for tweet_id in tweet_ids:
            count += 1
            print(str(count) + ": " + str(tweet_id))
            try:
                tweet = api.get_status(tweet_id, tweet_mode='extended')
                print("Success")
                json.dump(tweet._json, outfile)
                outfile.write('\n')
            except TweepyException as e:
                print("Fail")
                fails_dict[tweet_id] = e
                pass


# this will take about 30 min to run.
save_tweets_to_json()
```

### Issue #9: Duplicated url in single cell from `extended_urls` column

#### Define:
Duplicated urls are separated by comma, so we'll split them and remove duplicates.

#### Code

```python
def remove_duplicate_url(s):
    unique_urls = set(s.split(','))
    return ','.join(unique_urls)


twitter_archive_clean['expanded_urls'] = twitter_archive_clean.expanded_urls.apply(remove_duplicate_url)
# dropping duplicates just in case new one created
twitter_archive_clean.drop_duplicates("expanded_urls", inplace=True)
```

let's build a word cloud for the most frequent words in tweet posts.

```python
import string

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from wordcloud import WordCloud

def translate(text):
    """Removing unnecessary punctuation and tags."""
    translator = str.maketrans('', '', string.punctuation + string.digits)
    return text.translate(translator)


def remove_stopwords(words):
    """Removing stop words from `words` using the stopwords dictionary from ntlk"""
    en_stopwords = stopwords.words('english')
    en_stopwords = set(en_stopwords)
    words = text.lower().split()
    return [w for w in words if w not in en_stopwords]


def plot_word_cloud(words):
    wordcloud = WordCloud(collocations=False).generate(' '.join(words))
    plt.figure(figsize=(25,15))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.margins(x=0, y=0)
    plt.show()


def process_and_plot(text):
    text = translate(text)
    tokens = nltk.word_tokenize(text)
    words = remove_stopwords(tokens)
    cnt_fdist = nltk.FreqDist(words)
    common_20 = dict(cnt_fdist.most_common(100))
    plot_word_cloud(common_20)
```

```python
# concatenating all texts from twitter_archive
text = ' '.join(map(str, twitter_archive_clean['text'].values))
process_and_plot(text)
```
